```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 

arquivo = pd.read_csv('b1abb33c4f.csv')
df = pd.DataFrame(arquivo)
```

####  1) Na base do SINASC do DataSUS, faça um boxplot da idade da mãe.

```python
list(df.columns)
sns.boxplot(y=df['IDADEMAE'], data=df)
plt.ylabel('Idade Mãe')
```

####  2) Faça um boxplot do peso do bebê

```python
sns.boxplot(y=df['PESO'], data=df)
plt.ylabel('Peso Bebê (g)')
```

####  3) Faça um histograma da idade da mãe

```python
sns.displot(x=df['IDADEMAE'], data=df, alpha=0.25, bins = 20, element = 'step', kde = True)
plt.xlabel('Idade da Mãe')
plt.ylabel('Quantidade')
```

#### 4) Faça um histograma do peso do bebê

```python
sns.displot(x=df['PESO'], data=df, alpha=0.25, bins = 20, element = 'step', kde = True)
plt.xlabel('Peso do Bebê (g)')
plt.ylabel('Quantidade')
```
